```python
import pandas as pd
```

```python
movies_df = pd.read_csv("movies_metadata.csv")
top_df = pd.read_csv("blockbusters.csv")
cast_crew_df = pd.read_csv("credits.csv")
```

```python
#top_df = top_df.rename(columns={"title": "original_title"})
```

```python
merged = pd.merge(movies_df, top_df, on='title', how='inner')
```

```python
top_df.count
```

```python
merged.count
```

```python
merged['id'] = merged['id'].astype(int)
```

```python
top_cast = pd.merge(merged, cast_crew_df, on='id')
```

```python
top_cast.count
```

```python
import ast
import csv
```

```python
data_file = open('casts.csv', 'w')
csv_writer = csv.writer(data_file)
j_1 = ast.literal_eval(top_cast.cast[0])
csv_writer.writerow(["id", "title"] + list(j_1[0].keys()))
```

```python
for i, row in top_cast.iterrows():
    cast_list = ast.literal_eval(row['cast']) 
    for cast in cast_list:
        csv_writer.writerow([row['id'], row['title']] + list(cast.values()))
```

```python
data_file.close()
```
